```python
import urllib.request as req
import os
import json
import pandas as pd

serials = ["35241","35242","35243"]
dn = "baha/芙莉蓮"
def get_danmu(sn,dn,chapter=-1):
  if not os.path.exists(dn):
    os.makedirs(dn)
  url = f"https://api.gamer.com.tw/anime/v1/danmu.php?videoSn={sn}&geo=TW%2CHK"
  print("下載",url)

  f = req.urlopen(url)
  content = f.read()
  content = json.loads(content)
  danmus = content["data"]["danmu"]

  df = pd.DataFrame(danmus)
  df["ch"] = chapter
  df.to_csv(f"{dn}/{sn}.csv",encoding="utf-8")
  return df

get_danmu("35421","baha/芙莉蓮",30)


# for d in danmus:
#   print(d["userid"],d["text"])
```

```python
import glob
import pandas as pd
import os

dn = "baha/processed"
if not os.path.exists(dn):
  os.makedirs(dn)

total=[]

for fn in glob.glob("baha/*/*.csv"):
  df = pd.read_csv(fn,encoding="utf-8",index_col=0)
  total.append(df)
df = pd.concat(total)
df = df.sort_values(by="ch")
df = df.reset_index(drop=True)

# df["userid"].value_counts()
uids = df["userid"].unique()
for u in uids:
  fil = df["userid"] == u
  fp = f"{dn}/{u}.csv"
  df[fil].to_csv(fp,encoding="utf-8")


```
