```python
# 20000831(O) 20001331(X) 20000229(O) 20010229(X)
# is_date_valid()

def is_date_valid(d):
    day = d % 100
    remain = d // 100
    month = remain % 100
    year = remain // 100

    if year >= 1:
        if 12 >= month >= 1:
            month_days = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
            max_days = month_days[month-1]
            # if is_leap_year(year) == True
            if is_leap_year(year) and month == 2:
                max_days = max_days + 1
            if max_days >= day >= 1:
                return True
            else:
                return False
        else:
            return False
    else:
        return False

print(is_date_valid(20000229))
print(is_date_valid(20010229))
```

```python
# 20241001 2072...

def fully_squared_2(start=4499, max_count=10):

    result = []

    while True:

        d = start * start

        if is_date_valid(d):

            # list專屬功能: append

            # l = l.append(X) l.append(O)

            result.append(d)

        if len(result) == max_count:

            break

        start = start + 1

    return result

print(fully_squared_2(max_count=20))
```
