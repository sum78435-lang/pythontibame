```python
import urllib.request as req
import bs4 as bs

url = "https://tabelog.com/tw/kagoshima/rstLst/sushi/?SrtT=rt"
resp = req.urlopen(url)
content = resp.read()
html = bs.BeautifulSoup(content)
# html.find("a",{"class":"list-rst__rst-name-target"})
rst = html.find_all("div",{"class":"list-rst__body"})
for r in rst:
  name = r.find("a",{"class":"list-rst__rst-name-target"})
  print(name)
  name_href = name["href"]
  name_text = name.get_text()
  print(name_href)
  print(name_text)
  print("!")
```
